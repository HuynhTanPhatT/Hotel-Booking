# 3 - Data Quality Report

```python
def check_data_quality(df,col_name):
    quality_report = {
        'missing_values': df.isnull().sum(),
        'duplicated_values': df[col_name].duplicated().sum(),
        'unique_values' : len(df[col_name].unique()),
        'df_info': df.info()
    }
    return quality_report
```

# 4 - Data Cleaning

## 4.1 - Convert Data Types

```python
def convert_datetime(df,col_name):
    df[col_name] = pd.to_datetime(df[col_name])
def convert_types(df,col_name,type):
    df[col_name] = df[col_name].astype(type)
```

```python
convert_datetime(customer_table,col_name = 'created_at')
convert_datetime(booking_table,col_name = 'check_in')
convert_datetime(booking_table,col_name = 'check_out')
convert_datetime(booking_table, col_name = 'created_at')
convert_datetime(payment_table, col_name = 'payment_date')
```

## 4.2 - Create Columns

```python
customer_table['registeration_date'] = customer_table['created_at'].dt.strftime('%Y-%m-%d')
customer_table['year_month'] = customer_table['created_at'].dt.strftime('%Y-%m')
customer_table['year'] = customer_table['created_at'].dt.year
```

```python
# Create a column "month_name"
customer_table['month_name'] = customer_table['created_at'].dt.month_name()

#Create a list of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Sort month_name by month number
customer_table['month_name'] = pd.Categorical(customer_table['month_name'],
                                             categories= month_order,
                                             ordered=True)
```

```python
service_usage_table['price'] = round(service_usage_table['total_price'] / service_usage_table['quantity'])
convert_types(service_usage_table, col_name = 'price',type='Int64' )
```

```python
price_segment_by_room_type['Row Total']= price_segment_by_room_type.sum(numeric_only=True, axis=1) #Hng ngang
price_segment_by_room_type.loc['Column Total']= price_segment_by_room_type.sum(numeric_only=True,axis=0) #Hng dc
```

## 5.2.2 - The Most Used and Highest Revenue Services

```python
booking_service_usage = service_usage_table.merge(service_table,
                                 on = 'service_id', how = 'outer').drop(columns='price_y')

booking_service_usage = booking_service_usage.merge(booking_table[['booking_id','check_in','check_out']],
                                                   on = 'booking_id', how = 'left')
```

```python
booking_service_usage
```

```python
service_sales = booking_service_usage.groupby(['service_name','price_x'])[['quantity','total_price']].sum().reset_index().rename(columns={'total_price':'total_revenues',
                                                                                                                                         'price_x':'price_service'})
```

```python
sns.relplot(data = service_sales, x = 'price_service', y='total_revenues',
            hue = 'service_name', palette = 'Spectral',
            size = 'quantity',sizes = (50,300))
plt.show()
```

1. The scatter plot shows the postive correlation - the bigger number of sales and quantities(Typically, not always)
2. `Gym` with lower service have a lower overall revenue, although its number of sales performance are not less than other services.
3. `Room` Service recorded the most used service, however `SPA` make highest revenues

## 5.2.3 - The Average Guest Stay Duration by Month

```python
booking_service_usage['hotel_duration'] = (booking_service_usage['check_out'] - booking_service_usage['check_in']).dt.days
booking_service_usage['check_in_months'] = booking_service_usage['check_in'].dt.month
```

```python
avg_month_bookings = booking_service_usage.groupby('check_in_months')['hotel_duration'].mean().round(2)
```

```python
avg_month_bookings
```

```python
plt.bar(avg_month_bookings.index, avg_month_bookings.values,color='skyblue')

plt.title('Average Hotel Stay Duration by Month')
plt.xlabel('Month')
plt.ylabel('Hotel Durations (Avg Days)')
plt.ylim(7.0,8.0)
plt.xticks(range(1,13))

# Display Data
for i, v in enumerate(avg_month_bookings.values):
    plt.text(i + 1, v + 0.01, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
